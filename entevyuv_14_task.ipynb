```python
import pandas as pd
import seaborn as sns
```

```python
df1=pd.read_csv('/content/Rice Exporters 2019 - Trade Atlas Records - Jan.csv')
df2=pd.read_csv('/content/Rice Exporters 2020 - Jan.csv')
df3=pd.read_csv('/content/Rice Exporters 2021 - January.csv')
df4=pd.read_csv('/content/Rice Exporters 2022 - Jan.csv')
```

```python
df=pd.concat([df1,df2,df3,df4])
```

```python
pd.set_option('display.max_columns',30)
```

```python
df.drop(['FINAL COUNTRY','FINAL FOREIGN PORT','FINAL US PORT','US PORT2','US STATE','PREPORT','VOLUME UNIT','SHIPPER COUNTRY',
         'PORT OF DEPARTURE.1','Final Qty','NOTIFY PARTY','NOTIFY ADDRESS','MANUFACTORING COMPANY','BRAND NAME','PLACE OF DELIVERY',
         'INCOTERMS','PACKAGES UNIT','GROSS WEIGHT UNIT','NET WEIGHT UNIT','VOLUME','NO','HS CODE'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.describe()
```

```python
sns.scatterplot(df['QUANTITY'])
```

```python
outlier_col=['IMPORT VALUE FOB','IMPORT VALUE CIF','NET WEIGHT','GROSS WEIGHT','QUANTITY','NUMBER OF PACKAGES','CONTAINER COUNT','TOTAL TEUS']
for i in outlier_col:
  Q1=df[i].quantile(0.25)
  Q3=df[i].quantile(0.75)
  IQR=Q3-Q1
  LC=Q1-(1.5*IQR)
  UC=Q3+(1.5*IQR)
  df=df[(df[i]>=LC) & (df[i]<=UC)]
```

```python
df.shape
```

```python
df.head()
```

```python
#Importer/Exporter Overview
a=df.groupby('IMPORTER NAME').sum()[['QUANTITY']]
a=a.sort_values(by='QUANTITY',ascending=False).head(10)
a
```

```python
a.plot(kind='bar',ylabel='Quantity',title='Top 10 Importer based on Quantity')
```

```python
a=df.groupby('EXPORTER NAME').sum()[['QUANTITY']]
a=a.sort_values(by='QUANTITY',ascending=False).head(10)
a
```

```python
a.plot(kind='bar',ylabel='Quantity',title='Top 10 Exporter based on Quantity')
```
